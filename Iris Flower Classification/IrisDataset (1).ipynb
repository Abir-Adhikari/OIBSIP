```python
import numpy as np
import pandas as pd
import io
```

```python
df = pd.read_csv(io.BytesIO(uploaded['Iris.csv'])) #Read the dataset
df.head(5)
```

```python
#Returns tuple of shape of iris data classification
df.shape
```

```python
#Drop the Id Column
df = df.drop(columns=['Id'])
df.head()
```

```python
#Information
df.info()
```

```python
#Description
df.describe()
```

```python
#Transform Categorical Data Into Numeric Data 
df["Species"].replace({"Iris-setosa":1, "Iris-versicolor":2, "Iris-virginica":3},inplace = True)
df
```

```python
#Creating an array
arr = pd.DataFrame(df,columns=["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]).values
arr
```

```python
arr1 = df.Species.values.reshape(-1,1)
arr1
```

Importing Libraries for Machine Learning to train a model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split 
from sklearn import metrics
```

```python
#Train Test Split
x_train,x_test,y_train,y_test = train_test_split(arr,arr1,test_size=0.30, random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
kn=6
knclr=KNeighborsClassifier(kn)
```

```python
#Training the Model using K-Nearest Clustering
knclr.fit(x_train,y_train)
```

```python
y_pred=knclr.predict(x_test)
```

```python
metrics.accuracy_score(y_test,y_pred)*100
```
